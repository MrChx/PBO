```python
#Acha qilal addinutama p.nua
#T3122176
#Reg pagi A

class Buku:
    def __init__(self, isbn, judul, penulis, status, anggota=None, lama_dipinjam=0):
        self.isbn = isbn
        self.judul = judul
        self.penulis = penulis
        self.status = status
        self.anggota = anggota
        self.lama_dipinjam = lama_dipinjam

    def info_buku(self):
        info = f"ISBN: {self.isbn}\nJudul: {self.judul}\nPenulis: {self.penulis}\nStatus: {self.status}"
        if self.status == 'dipinjam':
            info += f"\nDipinjam oleh: {self.anggota}\nLama dipinjam: {self.lama_dipinjam} hari"
        return info

    def pinjam(self, anggota, lama_dipinjam):
        if self.status == 'tersedia':
            self.status = 'dipinjam'
            self.anggota = anggota
            self.lama_dipinjam = lama_dipinjam
            print(f"Buku '{self.judul}' berhasil dipinjam oleh {anggota}")
        else:
            print("Buku tidak tersedia untuk dipinjam")

    def kembalikan(self):
        if self.status == 'dipinjam':
            self.status = 'tersedia'
            self.anggota = None
            self.lama_dipinjam = 0
            print(f"Buku '{self.judul}' berhasil dikembalikan")
        else:
            print("Buku sudah tersedia")


class Perpustakaan:
    def __init__(self):
        self.daftar_buku = []

    def tambah_buku(self, buku):
        self.daftar_buku.append(buku)
        print(f"Buku '{buku.judul}' berhasil ditambahkan ke perpustakaan")

    def cari_buku(self, judul):
        for buku in self.daftar_buku:
            if buku.judul.lower() == judul.lower():
                return buku
        return None
    
    def pinjam_buku(self, judul, anggota, lama_dipinjam):
        buku = self.cari_buku(judul)
        if buku:
            buku.pinjam(anggota, lama_dipinjam)
        else:
            print("Buku tidak ditemukan")

    def kembalikan_buku(self, judul):
        buku = self.cari_buku(judul)
        if buku:
            buku.kembalikan()
        else:
            print("Buku tidak ditemukan")

if __name__ == "__main__":
    perpustakaan = Perpustakaan()

    perpustakaan.tambah_buku(Buku("0875298712", "Ancika", "Pidi Baiq", "tersedia"))
    perpustakaan.tambah_buku(Buku("92098713938", "Filosofi Teras", "Aris toteles", "tersedia"))
    perpustakaan.pinjam_buku("Ancika", "Acha Qilal", 5)
    perpustakaan.kembalikan_buku("Ancika")
    print("=========================================================")

    get_info = perpustakaan.cari_buku("Filosofi Teras")
    if get_info:
        print(get_info.info_buku())
```
